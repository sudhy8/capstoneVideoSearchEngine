```python
# CLIP
from transformers import CLIPProcessor

clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")


def clip_text_embeddings(text):
    inputs = clip_processor(text, return_tensors="pt", padding=True)
    input_tokens = {k: v for k, v in inputs.items()}

    return input_tokens["input_ids"]


# Example usage
text = "Image of a deer"
text_embeddings = clip_text_embeddings(text)
dee = text_embeddings
print(dee)

```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Normalize the text embedding vector
b_normalized = torch.nn.functional.normalize(text_embeds, p=2, dim=1)

# Get the image embeddings from the correct key in outputs
image_embeds = outputs['image_embeds']

# Normalize the image embedding vector
a_normalized = torch.nn.functional.normalize(image_embeds, p=2, dim=1)

# Calculate cosine similarity
similarity_score = cosine_similarity(a_normalized.cpu().detach().numpy(), b_normalized.cpu().detach().numpy())

print(f"Similarity score between the image and text vectors: {similarity_score[0][0]}")
```
